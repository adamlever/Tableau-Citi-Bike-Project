### Texture Analysis Project

Aim to read data from multiple excel files in a folder tree and to correlate in a single dataframe

```python
import os
import csv
import pandas as pd
```

```python
# Create empty lists to store file paths
file_paths_201401 = []
file_paths_201901 = []

# Loop through Data folder in project folder
for root, dirs, files in os.walk(r'C:\Users\alever\Desktop\Data Analytics Homework\Tableau\Data'):
    for file in files:
        if '.csv' in file:
            if root.endswith('2014 Trips'):
                print(file)
                path = os.path.join(root, file)
                print(path)
                file_paths_201401.append(path)
                
            if root.endswith('2019 Trips'):
                print(file)
                path = os.path.join(root, file)
                print(path)
                file_paths_201901.append(path)
```

```python
print(file_paths_201401)
print(file_paths_201901)
```

```python
# Create emptry dataframes to store 2014 data from .csv files
df201401 = pd.DataFrame()
df201901 = pd.DataFrame()

for f in file_paths_201401:
    df = pd.read_csv(f)
    df201401 = df201401.append(df,ignore_index=True)

# Save Dataframe to .csv
df201401.to_csv("2014_Trips.csv")

# Show dataframe
df201401
```

```python
# Create emptry dataframes to store 2014 data from .csv files
df201901 = pd.DataFrame()

for f in file_paths_201901:
    df = pd.read_csv(f)
    df201901 = df201901.append(df,ignore_index=True)

# Save Dataframe to .csv
df201901.to_csv("2019_Trips.csv")

# Show dataframe
df201901
```

```python
# Concatanate the 2014 and 2019 dataframes
Trips20142019 = pd.concat([df201401, df201901])

# Save Dataframe to .csv
Trips20142019.to_csv("Trips20142019.csv")

# Show new dataframe
Trips20142019
```
